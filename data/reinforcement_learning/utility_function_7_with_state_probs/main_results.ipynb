## Example 7: Plots of each server with different starting $\mu$'s

```python
def plot_utilities_and_weighted_rates_with_different_starting_rates(
    current_srv, utils_1, utils_2, utils_3, rates_1, rates_2, rates_3, state_probs_1, state_probs_2, state_probs_3
):
    _, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 7))

    ax1.plot([util[current_srv] for util in utils_1])
    ax1.plot([util[current_srv] for util in utils_2])
    ax1.plot([util[current_srv] for util in utils_3])

    # for srv in range(1, num_of_servers + 1):
    for all_rates, all_state_probs in zip(
        [rates_1, rates_2, rates_3], [state_probs_1, state_probs_2, state_probs_3]
    ):
        server_mean_weighted_rates = []
        for itr, _ in enumerate(all_rates):
            current_iteration_mean_weighted_rate = 0
            for state, current_rate in all_rates[itr][current_srv + 1].items():
                current_state_prob = all_state_probs[itr][state]
                if not np.isnan(current_state_prob) and state != (0, 0):
                    current_iteration_mean_weighted_rate += (
                        current_rate * current_state_prob
                    )
            server_mean_weighted_rates.append(current_iteration_mean_weighted_rate)
        ax2.plot(server_mean_weighted_rates)

    ax1.legend([f"$\mu_0 = 0.2$", f"$\mu_0 = 0.7$", f"$\mu_0 = 1.5$"])
    ax1.set_title(f"Utilities of server {current_srv + 1} over all iterations")
    ax2.legend([f"$\mu_0 = 0.2$", f"$\mu_0 = 0.7$", f"$\mu_0 = 1.5$"])
    ax2.set_title(f"Weighted mean rates of server {current_srv + 1} over all iterations");
```

```python
def plot_utilities_and_weighted_rates_using_ratio(
    utilities, all_rates, all_state_probs, num_of_servers, print_utils=False
):
    """
    Plot the utilities and the weighte mean rates of all servers over all
    iterations.

    Weighted rate = Rate at state (u,v) * Probability of state (u,v)
    """
    if print_utils:
        _, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 7))
        for srv in range(num_of_servers):
            ax1.plot([util[srv] for util in utilities])
        ax1.legend([f"Server {srv}" for srv in range(1, num_of_servers + 1)])
        ax1.set_title("Utilities of all servers over all iterations")
    else:
        _, ax2 = plt.subplots(1, 1, figsize=(20, 7))

    for srv in range(1, num_of_servers + 1):
        server_mean_weighted_rates = []
        for itr, _ in enumerate(all_rates):
            current_iteration_mean_weighted_rate = 0
            for state, current_rate in all_rates[itr][srv].items():
                current_state_prob = all_state_probs[itr][state]
                if not np.isnan(current_state_prob) and state != (0, 0):
                    current_iteration_mean_weighted_rate += (
                        (current_rate / max(all_rates[itr][srv].values())) * current_state_prob
                    )
            server_mean_weighted_rates.append(current_iteration_mean_weighted_rate)
        ax2.plot(server_mean_weighted_rates)

    ax2.legend([f"Server {srv}" for srv in range(1, num_of_servers + 1)])
    ax2.set_title("Weighted mean rate ratio (current rate / max rate) of all servers over all iterations")
```
